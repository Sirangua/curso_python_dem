# Procesamiento de imagenes

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

Las imagenes son matrices de pixeles

```python
img = np.array([[   0, 255,   0],   #   black, white, black, gray
                [  50, 200,  50],   
                [ 20, 128, 235]])  

plt.imshow(img, cmap='gray')
```

Los colores

```python
img = np.array([[[42,  47,   161], [ 255, 255,   0], [  0,   0, 255]],
                [[40,  190,   25], [ 25, 15,   10], [  110,   50, 255]],
                [[50,  190,   255], [ 255, 255,   0], [  0,   0, 255]]])  
plt.axis("off")
plt.imshow(img)
```

Abrir una imagen y mostrarla

```python
img = plt.imread("hombrefeliz.jpg")
plt.imshow (img)
```

Ver sus valores

```python
#print (img)
print (img[500,200])
```

Recortar

```python
print("Dimensiones imagen: ", np.shape(img))
```

```python
face = img[50:400,100:400,:]
plt.imsave('face.jpg',face)   # Save the image
plt.imshow(face)
```

Rotar una imagen

```python
face = np.rot90(face)
plt.imshow (face)
```

```python
face = np.fliplr(face)
face = np.flipud(face)
plt.imshow (face)
```

Convertir a escala de grises

```python
face_avg = np.dot(face[...,:3], [1/3,1/3,1/3])
plt.imshow(face_avg, cmap = 'gray')
```

```python
# Calculate regular average and weighted average
face_avg = np.dot(face[...,:3], [1/3,1/3,1/3])
face_wgt = np.dot(face[...,:3], [0.299,0.587,0.114])

# display images
fig, ax = plt.subplots(1,2)
ax[0].imshow(face_avg, cmap='gray')
ax[1].imshow(face_wgt, cmap='gray')
```

Hostograma

```python
plt.hist(face_wgt.ravel(),256,[0,255])
plt.title('Histogram')
plt.show()
```

```python
threshold = 150
_, img_bn = cv.threshold(face_wgt, threshold, 255, cv.THRESH_BINARY)

plt.imshow(img_bn, cmap='gray')
```
